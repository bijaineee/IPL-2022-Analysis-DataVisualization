```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

Basics of Matplotlib >

```python
x = np.linspace(0,5,11)
x
y = x **3 
y
```

Subplots in Matplotlib >

```python
plt.plot(x,y)
plt.title("Cube of Numbers")
plt.xlabel("X - axis")
plt.ylabel("Y - axis")
```

```python
plt.subplot(2,2,1)
plt.plot(x ** 5,y ** 5)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,2)
plt.plot(x**4,x**4)
```

Object oriented way >
Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the figsize and dpi keyword arguments.
figsize is a tuple of the width and height of the figure in inches dpi is the dots-per-inch (pixel per inch).

```python
fig = plt.figure(figsize=(20,5),dpi = 100)
axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
plt.title("Cube of Numbers")
plt.xlabel("X - axis")
plt.ylabel("Y - axis")
axis2 = fig.add_axes([0.3,0.5,.3,.3])
axis2.plot(x,y)
plt.title("Cube of Numbers")
plt.xlabel("X - axis")
plt.ylabel("Y - axis")
```

DISTRIBUTION PLOT >

```python
import seaborn as sns
import matplotlib.pyplot as plt 
```

We use distribution plots to visualize the distribution of quantitative data.

Let's discuss some plots that allow us to visualize the distribution of a data set. These plots are:

distplot/histplot
jointplot
pairplot
rugplot

```python
df = sns.load_dataset('tips')
df
```

```python
df['size'].unique()
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins = 20,kde = True)
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins = 20,kde = True)
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df,kind = 'scatter')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df,kind = 'reg')
```

```python
sns.pairplot(df,hue='size',palette='rainbow')
```

```python
sns.rugplot(df['tip'])
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df,kind = 'hex')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df,kind = 'kde')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df,kind = 'resid')
```

```python
sns.pairplot(df,hue='size',palette='rainbow')
```

```python
sns.rugplot(df['tip'])
```

CATEGORICAL DATA PLOTS >

Now let's discuss using seaborn to plot categorical data! There are a few main plot types for this:

boxplot >
violinplot >
stripplot >
swarmplot >
barplot >
countplot >

```python
df = sns.load_dataset("tips")
df
```

```python
sns.countplot(x=df['sex'])
```

```python
sns.countplot(x = df['sex'],hue = df['smoker'])
```

```python
sns.barplot(x = df['sex'],y = df['total_bill'])
```

```python
sns.barplot(x = df['sex'],y = df['tip'],estimator=np.sum)
```

```python
sns.boxplot(x= 'tip',y = 'day',data = df,palette = 'rainbow')
```

## boxplot and violinplot


```python
sns.violinplot(x = 'day',y = 'total_bill',data = df)
```

## stripplot and swarmplot

```python
sns.stripplot(x = 'day',y = 'total_bill',data = df)
```

```python
sns.swarmplot(x = 'day',y = 'total_bill',data = df)
```

```python
# swarmplot and violonplot together >

sns.violinplot(x= 'tip',y = 'day',data = df)
sns.swarmplot(x= 'tip',y = 'day',data = df)
```
